```python
# First, let's get all the hybrids in 2010
hb_10 = df_10[df_10['fuel'].str.contains('/')]
hb_10
```

```python
# create two copies of the 2010 hybrids dataframe
df1 = hb_10.copy()  # data on first fuel type of each hybrid vehicle
df2 = hb_10.copy()  # data on second fuel type of each hybrid vehicle

# Each one should look like this
df1
```

```python
# columns to split by "/"
split_columns = ['fuel', 'air_pollution_score', 'city_mpg', 'hwy_mpg', 'cmb_mpg', 'greenhouse_gas_score']

# apply split function to each column of each dataframe copy
for c in split_columns:
    df1[c] = df1[c].apply(lambda x: x.split("/")[0])
    df2[c] = df2[c].apply(lambda x: x.split("/")[1])
```

## How much have vehicle classes improved in fuel economy?

```python
#Let's look at the average fuel economy for each vehicle class for both years.
veh_10 = df_10.groupby('veh_class').cmb_mpg.mean()
veh_10
```

```python
veh_22 = df_22.groupby('veh_class').cmb_mpg.mean()
veh_22
```

```python
# how much they've increased by for each vehicle class
inc = veh_22 - veh_10
inc
```

```python
# only plot the classes that exist in both years
inc.dropna(inplace=True)
plt.subplots(figsize=(10,8))
plt.bar(inc.index, inc)
plt.title('Improvements in Fuel Economy from 2010 to 2022 by Vehicle Class')
plt.xlabel('Vehicle Class')
plt.ylabel('Increase in Average Combined MPG');
```
